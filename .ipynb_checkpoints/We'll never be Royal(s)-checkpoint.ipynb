# More Accurate Data Pre-Processing 

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

### Much easier way to obtain and manipulate the CSV file with Pandas

```python
df.head()
```

### Scatter Matrix lets us check all correlations instantaneously. Much more affective than going one-by-one

```python
# checking correlations 
from pandas.plotting import scatter_matrix
subset = df[['TotalConsmp','TempOutSide','Press_mm_hg','H_OutSide','Windspeed','Visibility']]
scatter_matrix(subset,figsize=(10,10)) #(subset, alpha=0.2, figsize=(6, 6), diagonal='kde')
plt.show()
```

```python
import seaborn as sns
```

### Data Visualization w/ Seaborn makes it much easier (and effective) to check correlations

```python
cols = ['TotalConsmp','TempOutSide','Press_mm_hg','H_OutSide','Windspeed','Visibility']
cm = np.corrcoef(subset.values.T)
sns.set(font_scale=1.5)
plt.figure(figsize=(10,10))
hm = sns.heatmap(cm,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws= {'size': 10},
                 yticklabels = cols,
                 xticklabels=cols
                )
plt.show()
```
